```python
def create_artist_table(artist): 
    t_a = []
    t_n = []
    t_p = []
    t_id = []
    for i in range(0,100,50):
        track_results = s.search(q='artist:' + artist, type='track', limit=50,offset=i)
        for i, t in enumerate(track_results['tracks']['items']):
            t_a.append(t['artists'][0]['name'])
            t_n.append(t['name'])
            t_id.append(t['id'])
            t_p.append(t['popularity'])

    temp_df = pd.DataFrame({'artist_name': t_a, 'track_name' : t_n, 'track_id' : t_id, 'track_popularity' : t_p})

    f = [] 
    for t in temp_df['track_id']:
        audio_features = s.audio_features(t)
        f.append(audio_features)
    f_df = pd.DataFrame(columns = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'type', 'id', 'uri', 'track_href', 'analysis_url', 'duration_ms', 'time_signature'])
    idx = 0
    for i in f: 
        s_data = f[idx][0]
        s_data_df = pd.DataFrame(s_data, [idx, ])
        f_df = pd.concat([f_df, s_data_df])
        idx += 1

    f_df['track_id'] = f_df['id']
    return_df = pd.merge(left = temp_df, right = f_df, on = 'track_id')
    return_df = return_df.drop(columns = ['id'])

    return return_df
```
